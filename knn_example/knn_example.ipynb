## KNN

- apple = 0
- banana = 1
- watermelon = 2

Import library to use

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generate data

```python
n: int = 500

apple_width: float = np.random.normal(6., 1.5, n)
apple_length: float = np.random.normal(6., 1.5, n)

banana_width: float = np.random.normal(4., 1.5, n)
banana_length: float = np.random.normal(8., 1.5, n)

watermelon_width: float = np.random.normal(20., 1.5, n)
watermelon_length: float = np.random.normal(20., 1.5, n)
```

```python
width = np.concatenate((apple_width, banana_width, watermelon_width))
length = np.concatenate((apple_length, banana_length, watermelon_length))
```

Plot data

```python
plt.scatter(apple_width, apple_length, c='red', marker='.')
plt.scatter(banana_width, banana_length, c='yellow', marker='.')
plt.scatter(watermelon_width, watermelon_length, c='green', marker='.')
plt.title('Apple and Banana And Watermelon')
plt.legend(['Apple', 'Banana', 'Watemelon'])
plt.xlabel('width')
plt.ylabel('length')

plt.show()
```

Define train data

```python
x_train: np.ndarray = np.array([width, length]).T
x_train
```

Define target

```python
y_apple: np.ndarray = np.zeros(n, dtype="int")
y_banana: np.ndarray = np.ones(n, dtype="int")
y_watermelon: np.ndarray = np.ones(n, dtype="int") * 2
```

```python
y_train = np.concatenate((y_apple, y_banana, y_watermelon))
y_train
```

Get nearest neighbors

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
new_fruit = np.array([20, 24])

distances: list = [euclidean_distance(new_fruit, x) for x in x_train]
distances
```

Define K

```python
k: int = 7
```

```python
nearest_neighbors = np.argsort(distances)[:k]
results = np.bincount(y_train[nearest_neighbors])
results
```

```python
result: int = np.argmax(results)
if result == 0:
    print("🍎")
elif result == 1:
    print("🍌")
else:
    print("🍉")
```
