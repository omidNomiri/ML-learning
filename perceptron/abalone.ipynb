```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from perceptron import Perceptron
```

```python
data = pd.read_csv("input/abalone.csv")
data.head()
```

```python
data['Sex'].replace(["F", "M", "I"], [0, 1, 2], inplace=True)
data.corr()
```

```python
X = np.array(data[['Length']])
Y = np.array(data[['Height']])

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, shuffle=True)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = Perceptron(.001, .01, 5)
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
```

```python
(ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))[1]
ax1.scatter(x_train, y_train, color="blue")
ax1.plot(x_train, y_pred, color="red")
ax1.set_xlabel("Length")
ax1.set_ylabel("Height")
ax1.set_title("Length And Height")

ax2.plot(model.losses, color="brown")
ax2.set_xlabel("Epochs")
ax2.set_ylabel("Loss")
ax2.set_title("Loss And Epochs With 5 Epochs")
plt.show()
```
