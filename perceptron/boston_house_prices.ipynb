```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from matplotlib.animation import FuncAnimation
from perceptron import Perceptron
```

```python
data = pd.read_csv("input/boston.csv")

data = data.drop(columns=["id"])
data.rename(columns={"medv": "price"}, inplace=True)
data.head()
```

```python
data.corr()
```

```python
X = np.array([data["rm"], data["zn"]])
X = X.T

Y = np.array(data[["price"]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, shuffle=True)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = Perceptron(.001, .1, 50)
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
```

```python
X1, X2 = x_train[:,0], x_train[:,1]
x1, x2 = np.meshgrid(X1, X2)
y_pred1, y_pred2 = model.w[0][0] * x1 + model.b[0][0], model.w[0][1] * x2 + model.b[0][1]

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.scatter(X1 , X2 , y_train,  color="green",alpha=0.3)
ax.plot_surface(x1, x2,  y_pred1 + y_pred2)
ax.set_xlabel("rm")
ax.set_ylabel("zn")
ax.set_zlabel("price")
plt.show()
```
