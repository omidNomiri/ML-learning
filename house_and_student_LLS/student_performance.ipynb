```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.model_selection import train_test_split
from LLS import LLS
```

```python
X = np.random.uniform(0, 12, 500)

Y = 2 * X + np.random.normal(0, 1, size=X.shape)
Y = np.interp(Y, (Y.min(), Y.max()), (0, 20))

X, Y = X.reshape(-1, 1), Y.reshape(-1, 1)
```

```python
plt.scatter(X, Y)
plt.title("Random Data Of Hours And Grades")
plt.xlabel("hours")
plt.ylabel("grade")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

```python
lls = LLS()
lls.fit(np.array(x_train), np.array(y_train))
w = lls.w
```

```python
plt.scatter(X, Y)
plt.plot(X, X * w, color="red")
plt.title("My LLS Of Hours And Grades")
plt.xlabel("hours")
plt.ylabel("grade")
plt.show()
```

```python
X, Y = x_train, y_train
X, Y = X.reshape(-1), Y.reshape(-1)
```

```python
slope, intercept, r, p, std_err = linregress(X, Y)

mymodel = []
for x in x_train:
    y = slope * x + intercept
    mymodel.append(y)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, mymodel, color="yellow")
plt.title("scipy LLS Of Hours And Grades")
plt.xlabel("hours")
plt.ylabel("grade")
plt.show()
```
